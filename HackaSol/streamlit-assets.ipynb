```python
import pandas as pd
```

```python
batters = pd.read_csv('predict/stream_X_test_batsmen_24.csv')
```

```python
bowlers = pd.read_csv('predict/streamlit_x_test_bowler_24.csv')
```

```python
# Group by 'batter' and sum the 'batsman_runs'
batter_runs = batters.groupby('batter')['total_runs'].sum().reset_index()

# Rename the column for better understanding
batter_runs = batter_runs.rename(columns={'total_runs': 'Total_Runs'})

# Sort the DataFrame by 'Total_Runs' in descending order
batter_runs_sorted = batter_runs.sort_values(by='Total_Runs', ascending=False)
```

```python
batter_runs_sorted
```

```python
batter_runs_sorted.to_csv('dataset/batters.csv')
```

```python
bowlers.info()
```

```python
# Group by 'batter' and sum the 'batsman_runs'
bowler_wickets = bowlers.groupby('bowler')['wickets'].sum().reset_index()

# Rename the column for better understanding
bowler_wickets = bowler_wickets.rename(columns={'wickets': 'Total_Wickets'})

# Sort the DataFrame by 'Total_Runs' in descending order
bowler_wickets_sorted = bowler_wickets.sort_values(by='Total_Wickets', ascending=False)
```
