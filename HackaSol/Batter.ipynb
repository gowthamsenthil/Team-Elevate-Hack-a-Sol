```python
import pandas as pd

# Function to calculate batter statistics
def calculate_batter_statistics(df, batter, opponent_team):
    # Filter data for the specific batter and opponent team
    filtered_df = df[(df['batter'] == batter) & (df['bowling_team'] == opponent_team)]
    
    # Calculate total runs scored
    total_runs = filtered_df['total_runs'].sum()
    
    # Calculate runs scored through boundaries (4s and 6s)
    boundary_runs = filtered_df[filtered_df['batsman_runs'].isin([4, 6])]['batsman_runs'].sum()
    
    # Calculate total number of balls faced
    balls_faced = filtered_df.shape[0]
    
    # Calculate boundary percentage
    boundary_percentage = (boundary_runs / total_runs) * 100 if total_runs > 0 else 0
    
    # Calculate number of centuries
    centuries = (filtered_df.groupby(['match_id', 'inning'])
                      .apply(lambda x: (x['total_runs'].sum() >= 100).sum())
                      .sum())
    
    # Calculate number of half-centuries
    half_centuries = (filtered_df.groupby(['match_id', 'inning'])
                         .apply(lambda x: (50 <= x['total_runs'].sum() < 100).sum())
                         .sum())
    
    # Calculate number of matches played
    matches_played = filtered_df['match_id'].nunique()
    
    # Calculate number of fours and sixes
    fours = filtered_df[filtered_df['batsman_runs'] == 4].shape[0]
    sixes = filtered_df[filtered_df['batsman_runs'] == 6].shape[0]
    
    return {
        'Total Runs Scored': total_runs,
        'Boundary Percentage': boundary_percentage,
        'Centuries': centuries,
        'Half-Centuries': half_centuries,
        'Matches Played': matches_played,
        'Balls Faced': balls_faced,
        'Fours': fours,
        'Sixes': sixes
    }

# Example usage
statistics = calculate_batter_statistics(data, 'Virat Kohli', 'Kolkata Knight Riders')
print(statistics)
```

```python
import pandas as pd

# Sample DataFrame
# data = pd.read_csv('your_data.csv')  # Load your DataFrame

def create_sorted_runs_dataframe(df):
    # Identify unique players who have batted
    batters = df['batter'].unique()
    
    # Identify unique players who have bowled
    bowlers = df['bowler'].unique()
    
    # Combine batters and bowlers to get a list of all unique players
    all_players = set(batters).union(set(bowlers))
    
    # Initialize a dictionary to store total runs for each player
    player_runs = {player: 0 for player in all_players}
    
    # Calculate total runs for batters
    batter_runs = df.groupby('batter')['batsman_runs'].sum().to_dict()
    
    # Update the dictionary with total runs for batters
    for player in batter_runs:
        player_runs[player] = batter_runs[player]
    
    # Convert the dictionary to a DataFrame
    player_runs_df = pd.DataFrame(list(player_runs.items()), columns=['Player', 'Total Runs'])
    
    # Sort the DataFrame by Total Runs in descending order
    sorted_player_runs = player_runs_df.sort_values(by='Total Runs', ascending=False).reset_index(drop=True)
    
    return sorted_player_runs

# Create the sorted DataFrame
sorted_runs_df = create_sorted_runs_dataframe(data)

# Print the sorted DataFrame
print(sorted_runs_df)

# Optionally, save to a CSV file
sorted_runs_df.to_csv('sorted_player_runs.csv', index=False)
```

```python
def calculate_bowler_statistics(df, bowler, opponent_team):
    # Filter the DataFrame for the specific bowler and opponent team
    filtered_df = df[(df['bowler'] == bowler) & (df['batting_team'] == opponent_team)]

    # Calculate total wickets taken
    total_wickets = filtered_df['is_wicket'].sum()

    # Calculate total runs conceded
    total_runs_conceded = filtered_df['total_runs'].sum()

    # Calculate total overs bowled (each unique over)
    total_overs_bowled = filtered_df.groupby(['match_id', 'over']).ngroup().nunique()

    # Calculate average runs per over (Economy Rate)
    economy_rate = (total_runs_conceded / total_overs_bowled) if total_overs_bowled > 0 else 0

    # Calculate total maidens (overs where 0 runs were conceded)
    maiden_overs = filtered_df.groupby(['match_id', 'over']).apply(lambda x: x['total_runs'].sum() == 0).sum()

    # Calculate five-wicket hauls (bowler takes 5 or more wickets in a match)
    five_wicket_hauls = filtered_df.groupby('match_id')['is_wicket'].sum().apply(lambda x: x >= 5).sum()

    return {
        'Total Wickets': total_wickets,
        'Total Runs Conceded': total_runs_conceded,
        'Economy Rate': f"{economy_rate:.2f}",
        'Total Overs Bowled': total_overs_bowled,
        'Maidens': maiden_overs,
        'Five-Wicket Hauls': five_wicket_hauls
    }
```

```python
import pandas as pd
import json

# List of valid teams
valid_teams = [
    'Mumbai Indians', 'Kolkata Knight Riders', 'Rajasthan Royals',
    'Chennai Super Kings', 'Sunrisers Hyderabad', 'Delhi Capitals',
    'Punjab Kings', 'Lucknow Super Giants', 'Gujarat Titans',
    'Royal Challengers Bengaluru'
]

# Function to get the list of teams each player (both batter and bowler) played against
def get_teams_played_by_each_player(df):
    # Filter the DataFrame to include only valid teams
    filtered_df = df[df['bowling_team'].isin(valid_teams)]
    
    # Get the teams played against by batters
    batter_teams = filtered_df.groupby('batter')['bowling_team'].unique().apply(list).to_dict()
    
    # Get the teams played against by bowlers
    bowler_teams = filtered_df.groupby('bowler')['batting_team'].unique().apply(list).to_dict()
    
    # Merge the results for batters and bowlers into a single dictionary
    players_teams = {}
    
    # Add batters to the dictionary
    for player, teams in batter_teams.items():
        if player not in players_teams:
            players_teams[player] = teams
    
    # Add bowlers to the dictionary or update if already exists
    for player, teams in bowler_teams.items():
        if player in players_teams:
            # Merge teams for both batting and bowling (remove duplicates)
            players_teams[player] = list(set(players_teams[player] + teams))
        else:
            players_teams[player] = teams
    
    # Return the final dictionary
    return players_teams

# Example DataFrame format, replace 'data' with your actual dataset
# df = pd.read_csv('path_to_your_file.csv')

# Call the function to get the teams played by each player
players_teams = get_teams_played_by_each_player(data)

# Save the result as a JSON file
with open('players_teams_played_against.json', 'w') as json_file:
    json.dump(players_teams, json_file, indent=4)

print("JSON file has been successfully created!")
```
